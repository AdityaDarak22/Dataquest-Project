### 2. Data Cleaning

- Since we will be performing data visualization for euro-us exchange rates we will be converting selective columns for data cleaning.
- but the same steps can be applied to other columns as well

```python
# Renaming the column to make it easier to understand
exchange_rates.rename(columns={"Period\\Unit:":"Time","[US dollar ]":"US_dollar"},inplace=True)
```

```python
exchange_rates["Time"] = pd.to_datetime(exchange_rates.loc[:,"Time"])
```

```python
exchange_rates.sort_values("Time",ascending=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates
```

```python
# creating a new df with time and us dollar columns
euro_to_dollar = exchange_rates.loc[:,['Time','US_dollar']]
```

```python
# Checking for different values in the us dollar dataset
euro_to_dollar.loc[:,"US_dollar"].value_counts()
```

```python
# There are 62 values in the dataset which has '-' as a value. we will be dropping all these values
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar']!='-']
```

```python
# converting the column into a numeric datatype
euro_to_dollar.loc[:,'US_dollar']= euro_to_dollar.loc[:,'US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

### 3.Rolling Mean

```python
# plotting the euro-dollar exchange_rate
plt.plot(euro_to_dollar.loc[:,"Time"],euro_to_dollar.loc[:,"US_dollar"])
plt.show()
```

- If we want to show the long term trends rather than showing daily variations we can make use of rolling mean method or moving averages method

```python
plt.figure(figsize=(9,6))
plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')
for i, roll in zip([2,3,4,5,6],[7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar.loc[:,"Time"],euro_to_dollar.loc[:,"US_dollar"].rolling(roll).mean())
    plt.title("Rolling Mean:"+ str(roll),weight='bold')
    
plt.tight_layout()

plt.show()
```

```python
#
euro_to_dollar['rolling_mean'] = euro_to_dollar.loc[:,"US_dollar"].rolling(30).mean()
```

### Coming up with an idea for story telling through data visualisation
- We show how the euro-dollar rate changed during the 2007-2008's financial crisis. We can also show the data for 2016 and 2009 for comparison. We can use a line plot.

```python
financial_years =euro_to_dollar[(euro_to_dollar.loc[:,"Time"].dt.year >=2006) &(euro_to_dollar.loc[:,"Time"].dt.year <=2009)]
```

```python
financial_years
```

```python
financial_years_7_8 =euro_to_dollar[(euro_to_dollar.loc[:,"Time"].dt.year >=2007) &(euro_to_dollar.loc[:,"Time"].dt.year <=2008)]
```

```python
financial_years_7_8
```

```python
# plotting two graphs to know the story
import matplotlib.style as style
style.use("fivethirtyeight")

fig, ax = plt.subplots(figsize=(8,3))
ax.plot(financial_years["Time"],financial_years["rolling_mean"],color='#A6D785')


ax.plot(financial_years_7_8.loc[:,"Time"],financial_years_7_8.loc[:,"rolling_mean"],color='red',linewidth=3)
plt.xticks(rotation=30)


ax.axvspan(xmin=13950,xmax=14100,alpha=0.5,color='grey')

ax.text(13000,1.65,s='Euro-USD rate peaked at 1.59 during 2007-2008 financial crisis',weight='bold',fontsize=14)
ax.text(13000,1.61,s="Euro-USD exchange rates between 2006 - 2010",fontsize=12)
ax.set_xticklabels([])
ax.set_yticklabels([])

x= 13100
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += 365
    
y=1.19

for row in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(13000, y ,row, alpha=0.6, fontsize=11)
    y+= 0.1

plt.show()
```

```python
bush_obama_trump = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]

pres_bush = bush_obama_trump[bush_obama_trump.loc[:,"Time"].dt.year<2009]

pres_obama = bush_obama_trump[(bush_obama_trump.loc[:,"Time"].dt.year >2009) & (bush_obama_trump.loc[:,"Time"].dt.year < 2017)]

pres_trump = bush_obama_trump[(bush_obama_trump.loc[:,"Time"].dt.year>2017) & (bush_obama_trump.loc[:,"Time"].dt.year<2021)]
```

```python
# exchange rates in different presidency rules
plt.figure(figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
for ax in [ax1,ax2,ax3,ax4]:
    ax.set_ylim(0.8,1.7)
    ax.set_yticks([1.0, 1.2, 1.4,1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5) 
    
ax1.plot(pres_bush['Time'], pres_bush['rolling_mean'],
        color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'],
                   alpha=0.3)
ax1.text(12300, 1.92, 'BUSH', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1.text(12000, 1.8, '(2001-2009)', weight='bold',
        alpha=0.3)

ax2.plot(pres_obama['Time'], pres_obama['rolling_mean'],
        color='#ffa500')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'],
                   alpha=0.3)
ax2.text(15100, 1.92, 'OBAMA', fontsize=18, weight='bold',
        color='#ffa500')
ax2.text(14950, 1.8, '(2009-2017)', weight='bold',
         alpha=0.3)

ax3.plot(pres_trump['Time'], pres_trump['rolling_mean'],
        color='#00B2EE')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'],
                   alpha=0.3)
ax3.text(17650, 1.92, 'TRUMP', fontsize=18, weight='bold',
        color='#00B2EE')
ax3.text(17560, 1.8, '(2017-2021)', weight='bold',
         alpha=0.3)

ax4.plot(bush_obama_trump['Time'], bush_obama_trump['rolling_mean'],
        color='red')
ax4.set_xticklabels([])

ax1.text(10800, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(10800, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)

plt.show()
```
